```python
## ============================= USE VSCODE CELL OR JUPYTER==============================
## +++++++++++++++++++++++++++ RUN EACH CELL ONE BY ONE  +++++++++++++++++++++++++++++++

import pandas as pd
import csv

## ---------- filling variable readContent with value in JSON file ----------
readContent = [
   {
       "user": {
           "id": 1,
           "name": "agus",
           "email": "agus@example.com"
       },
       "txs": [
           {
               "id": 1,
               "tx_date": "2020-01-01",
               "tx_amount": 10,
               "tx_type": "buy"
           },
           {
               "id": 3,
               "tx_date": "2020-02-01",
               "tx_amount": 10,
               "tx_type": "buy"
           },
           {
               "id": 5,
               "tx_date": "2020-03-01",
               "tx_amount": 20,
               "tx_type": "sell"
           }
       ]
   },
   {
       "user": {
           "id": 2,
           "name": "asep",
           "email": "asep@example.com"
       },
       "txs": [
           {
               "id": 2,
               "tx_date": "2020-01-01",
               "tx_amount": 10,
               "tx_type": "sell"
           },
           {
               "id": 4,
               "tx_date": "2020-02-01",
               "tx_amount": 10,
               "tx_type": "sell"
           },
           {
               "id": 6,
               "tx_date": "2020-03-01",
               "tx_amount": 20,
               "tx_type": "buy"
           }
       ]
   }
]

## ---------- filling variable readContent with value in JSON file ----------
```

```python
## ----------------JSON to CSV file (CSV_1.csv)-----------------
fields = ['id', 'name', 'email']
user_id = []
for i in range (0, len(readContent)):
    part1txs=readContent[i]['user']
    user_id.append(list(part1txs.values())[0])

tx_id = []
for i in range(0, len(readContent)):
    parttxs = readContent[i]['txs']
    for i in range(0,len(parttxs)):
        tx_id.append(parttxs[i]['id'])

temp = []
for i in range (0, len(user_id)):
    test = i
    for i in range (0, len(readContent[test]['txs'])):
        temp.append(user_id[test])

dict = {'user_id': temp, 'tx_id': tx_id}
df = pd.DataFrame(dict)
df.to_csv('CSV_1.csv', index=False)
## ----------------JSON to CSV file (CSV_1.csv)-----------------
```

```python
## ----------------JSON to CSV file (CSV_2.csv)-----------------
fields = ['Id', 'name', 'email']
rows = []
for i in range (0, len(readContent)):
    part1txs=readContent[i]['user']
    rows.append(list(part1txs.values()))

with open('CSV_2.csv','w', newline='') as csvfile:
    csvwriter = csv.writer(csvfile)
    csvwriter.writerow(fields)
    csvwriter.writerows(rows)
## ----------------JSON to CSV file (CSV_2.csv)-----------------
```

```python
## ----------------JSON to CSV file (CSV_3.csv)-----------------

id0 = []
date0 = []
amount0 = []
type0 = []

for i in range(0, 2, len(readContent)):
    parttxs = readContent[i]['txs']
    parttxs1 = readContent[i+1]['txs']
    for i in range(0,len(parttxs)):
        id0.append(parttxs[i]['id'])
        id0.append(parttxs1[i]['id'])
        date0.append(parttxs[i]['tx_date'])
        date0.append(parttxs1[i]['tx_date'])
        amount0.append(parttxs[i]['tx_amount'])
        amount0.append(parttxs1[i]['tx_amount'])
        type0.append(parttxs[i]['tx_type'])
        type0.append(parttxs1[i]['tx_type'])

dict = {'Id': id0, 'tx_date': date0, 'tx_amount': amount0, 'tx_type': type0}
df = pd.DataFrame(dict)
df.to_csv('CSV_3.csv', index=False)
## ----------------JSON to CSV file (CSV_3.csv)-----------------
```
