```python
## =================== USE VSCODE OR JUPYTER====================
## +++++++++++++++++ RUN EACH CELL ONE BY ONE +++++++++++++++++++++

## ----------------Database Initialization-----------------
import sqlite3
from sqlite3 import Error
def location():
    return (r"c:\sqlite\labPython1.db") ##<<<<<<<<<<<<<< INITIALIZATION DATABASE TARGET <<<<<<<<<<<<<<

def create_connection(db_file):
    conn = None
    try:
        conn = sqlite3.connect(db_file)
    except Error as e:
        print(e)
    finally:
        if conn:
            conn.close()

if __name__ == '__main__':
    create_connection(location()) 

## ================Database Initialization===================
```

```python
## ----------------Table Header Initialization-----------------
def create_connection(db_file):
    conn = None
    try:
        conn = sqlite3.connect(db_file)
        return conn
    except Error as e:
        print(e)
    return conn

def create_table(conn, create_table_sql):
    try:
        cur = conn.cursor()
        cur.execute(create_table_sql)
    except Error as e:
        print(e)

def main():
    database = location()

    create_table_header = """CREATE TABLE IF NOT EXISTS transactions (
                                    trx_id integer PRIMARY KEY,
                                    user_id integer NOT NULL,
                                    insert_time timestamp NOT NULL,
                                    tx_amount float NOT NULL,
                                    tx_type text NOT NULL,
                                    status text NOT NULL,
                                    FOREIGN KEY (user_id) REFERENCES users (id)
                                );"""
    conn = create_connection(database)
    if conn is not None:
        create_table(conn, create_table_header)
    else:
        print("Error! cannot create the database connection.")


if __name__ == '__main__':
    main()
## ================Table Header Initialization===================
```

```python
## ------------------Query to show [MAX(column_name)]---------------------
import pandas as pd 

def main():
    database = location()
    conn = create_connection(database)
    with conn:
        query = """SELECT * FROM transactions
                GROUP BY user_id
                HAVING MAX(insert_time)"""
    
    df = pd.read_sql_query(query, conn)
    conn.close()
    print(df)

if __name__ == '__main__':
    main()

## ================Query to show [MAX(column_name)]===================
```
